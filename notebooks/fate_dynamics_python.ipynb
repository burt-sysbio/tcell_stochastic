```python
# load packages
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read output from julia simulation
df = pd.read_csv("../output/teststoring.csv")
df2 = df.melt(id_vars = ["time"])
g = sns.relplot(data = df2, x = "time", y = "value", hue = "variable", kind = "scatter")
```
